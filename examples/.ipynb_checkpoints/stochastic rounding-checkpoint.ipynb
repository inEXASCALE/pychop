```python
import pychop
from pychop import Chop
import numpy as np
import time
np.random.seed(0)
import torch
X = torch.randn(5000, 5000) 
pychop.backend('torch', 1) # Specify different backends, e.g., jax and torch
backend = pychop.get_backend() # you can also get the current backend via .get_backend()
# One can also specify 'auto', the pychop will automatically detect the types,
# but speed will be degraded. 
 
ch = Chop(exp_bits=5, sig_bits=10, rmode=3) # half precision

st = time.time()
X_q = ch(X)
et = time.time()

print(X_q[:10, 0], f"runtime{et - st}")
```

```python
ch = Chop(exp_bits=5, sig_bits=10, rmode=6) # half precision

st = time.time()
X_q = ch(X)
et = time.time()

print(X_q[:10, 0], f"runtime:{et - st}")
```
